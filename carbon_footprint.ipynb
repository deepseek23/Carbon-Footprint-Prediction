```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
encoded = mlb.fit_transform(df['Recycling'])
```

```python
recycle_df = pd.DataFrame(encoded, columns=mlb.classes_, index=df.index)
```

```python
df = pd.concat([df.drop(columns=['Recycling']), recycle_df], axis=1)
```
