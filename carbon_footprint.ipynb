```python
from scipy.stats import chi2_contingency

# Get all categorical (binary) features
cat_features = [
    'Sex', 'Body Type_obese', 'Body Type_overweight', 'Body Type_underweight',
    'Diet_pescatarian', 'Diet_vegan', 'Diet_vegetarian',
    'How Often Shower_less frequently', 'How Often Shower_more frequently',
    'How Often Shower_twice a day', 'Heating Energy Source_electricity',
    'Heating Energy Source_natural gas', 'Heating Energy Source_wood',
    'Transport_public', 'Transport_walk/bicycle', 'Social Activity_often',
    'Social Activity_sometimes', 'Frequency of Traveling by Air_never',
    'Frequency of Traveling by Air_rarely',
    'Frequency of Traveling by Air_very frequently', 'Waste Bag Size_large',
    'Waste Bag Size_medium', 'Waste Bag Size_small', 'Vehicle Type_electric',
    'Vehicle Type_hybrid', 'Vehicle Type_lpg', 'Vehicle Type_petrol',
    'Vehicle Type_unknown', 'Energy efficiency_Sometimes',
    'Energy efficiency_Yes', 'Glass', 'Metal', 'Paper', 'Plastic',
    'Airfryer', 'Grill', 'Microwave', 'Oven', 'Stove'
]

alpha = 0.05

# Create bins for the target variable (CarbonEmission)
df['carbon_bin'] = pd.qcut(df['CarbonEmission'], q=4, labels=False)
chi2_results = {}

for col in cat_features:
    if col not in df.columns:
        print(f'Skipping missing column: {col}')
        continue
    contingency = pd.crosstab(df[col], df['carbon_bin'])
    # Need at least 2 rows and 2 columns to run chi2
    if contingency.size == 0 or contingency.shape[0] < 2 or contingency.shape[1] < 2:
        print(f'Skipping column with insufficient data for chi2: {col}')
        continue
    try:
        chi2_stat, p_val, _, _ = chi2_contingency(contingency)
    except Exception as e:
        print(f'Chi2 failed for {col}: {e}')
        continue
    decision = 'Reject Null (Keep Feature)' if p_val < alpha else 'Accept Null (Drop Feature)'
    chi2_results[col] = {
        'chi2_statistic': chi2_stat,
        'p_value': p_val,
        'Decision': decision
    }

chi2_df = pd.DataFrame(chi2_results).T
chi2_df = chi2_df.sort_values(by='p_value')
chi2_df
```

```python
# Get features to drop (those with p-value >= 0.05)
features_to_drop = chi2_df[chi2_df['Decision'] == 'Accept Null (Drop Feature)'].index.tolist()
print(f"Features to drop: {features_to_drop}")
print(f"\nTotal features to drop: {len(features_to_drop)}")

# Drop the features and also drop the temporary carbon_bin column
df_filtered = df.drop(columns=features_to_drop + ['carbon_bin'])
print(f"\nOriginal shape: {df.shape}")
print(f"Filtered shape: {df_filtered.shape}")
```

```python
# Load Optuna trials from optuna_study.db into MLflow for visualization
import json
import tempfile
from pathlib import Path

import mlflow
import optuna
from mlflow.tracking import MlflowClient
from optuna.trial import TrialState


def find_project_dir() -> Path:
    """Locate the folder that contains optuna_study.db."""
    candidates = [
        Path.cwd(),
        Path.cwd() / "Carbon footprint",
        Path(r"c:\Users\hp\Desktop\AIProjects\ml projects\Carbon footprint"),
    ]
    for path in candidates:
        if (path / "optuna_study.db").exists():
            return path.resolve()
    raise FileNotFoundError(
        "optuna_study.db not found. Run this notebook from the Carbon footprint folder."
    )


def to_str(value):
    try:
        return str(value)
    except Exception:
        return repr(value)


PROJECT_DIR = find_project_dir()
OPTUNA_DB = PROJECT_DIR / "optuna_study.db"
MLFLOW_DB = PROJECT_DIR / "mlflow.db"

storage_url = f"sqlite:///{OPTUNA_DB.as_posix()}"
mlflow_uri = f"sqlite:///{MLFLOW_DB.as_posix()}"
study_name = "carbon_emission_xgb"
exp_name = "CarbonEmission_Optuna_Trials"

mlflow.set_tracking_uri(mlflow_uri)
mlflow.set_experiment(exp_name)

study = optuna.load_study(study_name=study_name, storage=storage_url)
client = MlflowClient()
exp = mlflow.get_experiment_by_name(exp_name)
exp_id = exp.experiment_id

print(f"Project dir : {PROJECT_DIR}")
print(f"Loaded study: {study_name} with {len(study.trials)} trials")
print(f"Best CV R²  : {study.best_value:.4f}")
print(f"MLflow DB   : {MLFLOW_DB}")

imported = 0
skipped = 0

for trial in study.trials:
    trial_num = trial.number
    existing = client.search_runs(
        [exp_id],
        filter_string=f"tags.optuna_trial_number = '{trial_num}'",
        max_results=1,
    )
    if existing:
        skipped += 1
        continue

    params = {k: to_str(v) for k, v in trial.params.items()}

    with mlflow.start_run(experiment_id=exp_id, run_name=f"optuna_trial_{trial_num}"):
        if params:
            mlflow.log_params(params)

        if trial.value is not None:
            mlflow.log_metric("cv_r2", float(trial.value))
            mlflow.log_metric("objective", float(trial.value))

        mlflow.set_tag("optuna_trial_number", to_str(trial_num))
        mlflow.set_tag("optuna_state", to_str(trial.state.name))
        mlflow.set_tag("optuna_study", study_name)

        for key, value in trial.user_attrs.items():
            mlflow.set_tag(f"user_{key}", to_str(value))

        for step, value in trial.intermediate_values.items():
            mlflow.log_metric(f"intermediate_{step}", float(value))

        meta = {
            "number": trial_num,
            "state": trial.state.name,
            "value": trial.value,
            "params": params,
        }
        with tempfile.NamedTemporaryFile(mode="w", delete=False, suffix=".json") as tmp:
            json.dump(meta, tmp, indent=2)
            tmp_path = tmp.name
        mlflow.log_artifact(tmp_path, artifact_path="optuna_meta")

    imported += 1

total_runs = len(client.search_runs([exp_id]))
print(f"Imported {imported} new trials, skipped {skipped} existing trials")
print(f"MLflow experiment '{exp_name}' now has {total_runs} runs")
print("Next: run the evaluation cell, then launch MLflow UI from the last cell.")
```

```python
def objective(trial):

    params = {
        "n_estimators": trial.suggest_int("n_estimators", 100, 1000),
        "max_depth": trial.suggest_int("max_depth", 3, 15),
        "learning_rate": trial.suggest_float(
            "learning_rate", 0.001, 0.3, log=True
        ),
        "subsample": trial.suggest_float("subsample", 0.5, 1.0),
        "colsample_bytree": trial.suggest_float(
            "colsample_bytree", 0.5, 1.0
        ),
        "gamma": trial.suggest_float("gamma", 0, 10),
        "min_child_weight": trial.suggest_int(
            "min_child_weight", 1, 10
        ),
        "reg_alpha": trial.suggest_float(
            "reg_alpha", 1e-5, 10, log=True
        ),
        "reg_lambda": trial.suggest_float(
            "reg_lambda", 1e-5, 10, log=True
        ),
        "random_state": 42,
        "objective": "reg:squarederror",
        "tree_method": "hist"
    }

    model = xgb.XGBRegressor(**params)

    score = cross_val_score(
        model,
        X_train,
        y_train,
        cv=5,
        scoring="r2",
        n_jobs=-1
    ).mean()

    return score


# =========================
# Run Optuna using SQLite study DB (loads existing study or creates one)
# =========================

storage_url = 'sqlite:///optuna_study.db'
study_name = 'carbon_emission_xgb'
desired_total_trials = 100

# Try loading an existing study, otherwise create it
try:
    study = optuna.load_study(study_name=study_name, storage=storage_url)
    print(f'Loaded existing study: {study_name} with {len(study.trials)} trials')
except optuna.exceptions.StudyNotFound:
    study = optuna.create_study(study_name=study_name, direction='maximize', storage=storage_url)
    print(f'Created new study: {study_name}')

# Run only the remaining trials needed to reach desired_total_trials
existing = len(study.trials)
if existing < desired_total_trials:
    remaining = desired_total_trials - existing
    print(f'Running {remaining} additional trials (to reach {desired_total_trials})')
    study.optimize(objective, n_trials=remaining)
else:
    print(f'Study already has {existing} trials; skipping optimization')

print('Best R²:', study.best_value)
print('Best Params:')
print(study.best_params)

best_params = study.best_params

# =========================
# MLflow Tracking
# =========================
```

```python
from pathlib import Path

mlflow_db = Path(r"c:\Users\hp\Desktop\AIProjects\ml projects\Carbon footprint") / "mlflow.db"
mlflow.set_tracking_uri(f"sqlite:///{mlflow_db.as_posix()}")
mlflow.set_experiment("CarbonEmission_XGBoost_Optuna")

with mlflow.start_run(run_name=f"best_trial_{study.best_trial.number}"):

    # Log Optuna metadata
    mlflow.log_param("optuna_trial_number", study.best_trial.number)
    mlflow.log_param("optuna_best_score", study.best_trial.value)

    model = xgb.XGBRegressor(
        **best_params,
        random_state=42,
        objective="reg:squarederror"
    )

    model.fit(X_train, y_train)

    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)

    # Train metrics
    train_r2 = r2_score(y_train, train_pred)
    train_rmse = mean_squared_error(
        y_train,
        train_pred
    ) ** 0.5
    train_mae = mean_absolute_error(
        y_train,
        train_pred
    )

    # Test metrics
    test_r2 = r2_score(y_test, test_pred)
    test_rmse = mean_squared_error(
        y_test,
        test_pred
    ) ** 0.5
    test_mae = mean_absolute_error(
        y_test,
        test_pred
    )

    # Log best Optuna params
    mlflow.log_params(best_params)

    # Log metrics
    mlflow.log_metrics({
        "train_r2": train_r2,
        "test_r2": test_r2,
        "train_rmse": train_rmse,
        "test_rmse": test_rmse,
        "train_mae": train_mae,
        "test_mae": test_mae
    })

    # Log model
    mlflow.xgboost.log_model(
        model,
        artifact_path="xgboost_model"
    )

    print(f"Train R² : {train_r2:.4f}")
    print(f"Test R²  : {test_r2:.4f}")
```

```python
# Evaluate each completed Optuna trial on train/test and log metrics to MLflow
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from optuna.trial import TrialState


def get_mlflow_run_for_trial(trial_number: int):
    runs = client.search_runs(
        [exp_id],
        filter_string=f"tags.optuna_trial_number = '{trial_number}'",
        max_results=1,
    )
    return runs[0] if runs else None


def build_model_from_trial(trial):
    params = trial.params.copy()
    params.update(
        random_state=42,
        objective="reg:squarederror",
        tree_method="hist",
    )
    return xgb.XGBRegressor(**params)


evaluated = 0
skipped_eval = 0

for trial in study.trials:
    if trial.state != TrialState.COMPLETE or trial.value is None:
        skipped_eval += 1
        continue

    run = get_mlflow_run_for_trial(trial.number)
    if run is None:
        print(f"No MLflow run found for trial {trial.number}; run the import cell first.")
        skipped_eval += 1
        continue

    if "test_r2" in run.data.metrics:
        skipped_eval += 1
        continue

    model = build_model_from_trial(trial)
    model.fit(X_train, y_train)

    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)

    with mlflow.start_run(run_id=run.info.run_id):
        mlflow.log_metrics(
            {
                "train_r2": r2_score(y_train, train_pred),
                "test_r2": r2_score(y_test, test_pred),
                "train_rmse": mean_squared_error(y_train, train_pred) ** 0.5,
                "test_rmse": mean_squared_error(y_test, test_pred) ** 0.5,
                "train_mae": mean_absolute_error(y_train, train_pred),
                "test_mae": mean_absolute_error(y_test, test_pred),
            }
        )

    evaluated += 1
    if evaluated % 10 == 0:
        print(f"Evaluated {evaluated} trials...")

print(f"Evaluated {evaluated} trials, skipped {skipped_eval}")

best_run = get_mlflow_run_for_trial(study.best_trial.number)
if best_run:
    metrics = best_run.data.metrics
    print("\nBest trial summary")
    print(f"  Trial #: {study.best_trial.number}")
    print(f"  CV R²  : {metrics.get('cv_r2', study.best_value):.4f}")
    print(f"  Test R²: {metrics.get('test_r2', 'not evaluated yet')}")
```
